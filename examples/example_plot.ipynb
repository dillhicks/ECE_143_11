```python
import pandas as pd


data = pd.read_csv("Forbes Global 2000, 2008 - 2019.csv")
data_2019 = data.loc[data['year'] == 2019]
```

```python
import plotly.express as px
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected=True)   


fig = px.scatter_3d(data_2019, x="PCA_1", y="PCA_2", z="PCA_3", color="pca_cluster", hover_data=[
            "PCA_1", "PCA_2", 
            "PCA_3", "Company"])
fig.show()
```

```python
fig = px.scatter_3d(data_2019, x="PCAtfidf_1", y="PCAtfidf_2", z="PCAtfidf_3", color="tfidf_cluster", hover_data=[
            "PCA_1", "PCA_2", 
            "PCA_3", "Company"
        ])
fig.show()
```

```python
data_cluster1 = data_2019.loc[data_2019['tfidf_cluster'] == 1]

fig = px.scatter_3d(data_cluster1, x="PCAtfidf_1", y="PCAtfidf_2", z="PCAtfidf_3", color="tfidf_cluster", hover_data=[
            "PCA_1", "PCA_2", 
            "PCA_3", "Company"
        ])
fig.show()
```
